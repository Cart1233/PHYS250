# Problem 1
(5 points)

Write a function for the following equation:

$$ y(x) = 0.9 x^4 - 12x^2 -5x $$

The input to the function is x and it returns y.

1. Test that your function returns the correct values for $y(-3)$ and $y(5)$.
1. Use the function to make a plot of $y(x)$ for $-4 \le x \le 4$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Your answer here
""" the equations defines an equation that takes in x.
The next line is the given equation that is then followed by a return statement in order for the equation to work. """
def equation(x):
  y=.9*(x**4)-12*(x**2)-5*x
  return y
```

```python
""" These staements provide proof that the equation works """
test1=equation(-3)
test2=equation(5)
print(test1)
print(test2)
```

```python
"""plotting code"""

x_val=np.linspace(-4,4,9)
y_val=equation(x_val)

plt.plot(x_val,y_val)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y(x)=.9x^4-12x^2-5x")
plt.legend()
```

# Problem 2
(5 points)

Write a function that converts polar coordinates to cartesian coordinates.  The function takes $r,\theta$ and returns $x$ and $y$, where $\theta$ is in radians.

1. Test that your function returns the correct values for $r=1,\theta=0$ and $r=1,\theta=\pi$.
1. Use the function to make a plot of the $x$ and $y$ values that correspond to:

$$ r: 0 \le r \le 1$$
$$ \theta: 0 \le \theta \le 2\pi $$

You should first create the $r$ and $\theta$ arrays, using 100 elements in each array.

3. Recreate the figure above using the plt.scatter command, and color code each point by its index in the array.


```python
# Your answer here

def polar2cartesian(r, theta):
  """Function takes in theta and r"""
  x=r*(np.cos(theta))
  y=r*(np.sin(theta))
  """these are the functions that convert r and theta to x and y, respectively"""

  return x, y
```

```python
#test functions

test1=polar2cartesian(1,0)
test2=polar2cartesian(1,np.pi)

print(test1)
print(test2)
```

```python
# x and y plot

r_val=np.linspace(0,1,100)
theta_val=np.linspace(0,2*np.pi,100)
"""This assigns values to r and theta"""

cartesianx,cartesiany=polar2cartesian(r_val,theta_val)
"""This converts r and theta to cartesian coordinates"""

plt.plot(cartesianx,cartesiany)
plt.xlabel("Cartesian Coordinates of X")
plt.ylabel("Cartesian Coordinates of y")
plt.title("Cartesian Polar Coordinates ")
```

```python
#scatter plot

plt.scatter(cartesianx,cartesiany, c=range(len(cartesianx)))
plt.xlabel("Cartesian Coordinates of X")
plt.ylabel("Cartesian Coordinates of y")
plt.title("Cartesian Polar Coordinates ")
```

# Problem 3

Write a function called guess_number that takes in a number between one and 10 and compares the input with a random integer (using np.random.randint(1,high=10) to select the random number).  



*   Call a different random number each time the function is called.
*   Make sure the input number is between 1 and 10.  If not, print a warning message.
*   If the input number and the random number are the same, print "You win!".
*   If the input number is 7, print "Lucky 7s always win!".
*   Otherwise, print a message to let the user know their guess was wrong, and let them know what the random number was.
*   Test your function to make sure it works.



```python
# Your function should go here

def guess_number(num):
  """Function takes in 1 variable num"""
  randnum=np.random.randint(1,high=10)
  """This creates a random number"""
  if num<1 or num>10:
    print("The number is not between 1 and 10")
    """This is prints a statement if the number is not between 1 and 10"""
  elif randnum==num:
    print("You win!")
    """This is for if the input number and random number are the same"""
  elif num==7:
    print("Lucky 7s always win!")
    """If the input was 7 this will print"""
  else:
    print(f"Wrong guess! The answer was{randnum}")
```

```python
# Put your test code here

test1=guess_number(11)
print(test1)
"""tests number outside range"""
test2=guess_number(9)
print(test2)
"""Tests number in the range"""
test3=guess_number(7)
print(test3)
"""Tests the print statement for when input is 7"""
```
